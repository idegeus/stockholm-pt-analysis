```python
from google.transit import gtfs_realtime_pb2
import os
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
import contextily as cx
import matplotlib.pyplot as plt
from glob import glob
from tqdm.notebook import tqdm
import seaborn as sns
```

```python
# Trips and routes.
trips = pd.read_csv('./data/2-working-data/all-trips.csv', index_col=0)
trips.trip_id = trips.trip_id.astype(str)
trips = trips.drop(columns=['service_id', 'trip_headsign', 'agency_id', 'route_type', 'route_long_name'])

# Filter only Tvärbanan
trips_selected = trips[(trips.route_short_name == '2') == True]
trips_selected.sample(5)
```

```python
# Stops selected for Tvärbanan
stops = pd.read_csv('./data/2-working-data/route-2-stops.csv', index_col=0)
stops.stop_id = stops.stop_id.astype(str)

# Change to GeoDataFrame
stops['geometry'] = stops.apply(lambda x: Point(x.stop_lon, x.stop_lat), axis=1)
stops = stops.drop(columns=['location_type', 'stop_lat', 'stop_lon', 'platform_code', 'parent_station'])
stops = gpd.GeoDataFrame(stops, crs='EPSG:4326')

# Plot out and show stops
ax = stops.plot()
cx.add_basemap(ax, crs=stops.crs)

stops.head()
```

```python
# Read all CSV's in the dataframe
tu_df = pd.concat([
    pd.read_csv(csv_file, index_col=0) 
    for csv_file in glob('./data/10-cleaned-data-camille/*.csv')])
tu_df.trip_id = tu_df.trip_id.astype(str)
tu_df.stop_id = tu_df.stop_id.astype(str)
tu_df.timestamp    = pd.to_datetime(tu_df.timestamp, unit='s')
tu_df.arrival_ts   = pd.to_datetime(tu_df.arrival_ts, unit='s')
tu_df.departure_ts = pd.to_datetime(tu_df.departure_ts, unit='s')

tu_df.head()
```

```python
# Merge trips, routes and stop-data into TripUpdate_df. 
tu_df_merged = tu_df.merge(trips, how='left', on='trip_id')
tu_df_merged = tu_df_merged.merge(stops, how='left', on='stop_id')
tu_df_merged[['timestamp', 'trip_id', 'stop_id', 'arrival_delay', 
              'departure_delay', 'route_desc', 'stop_name']].tail()
```

```python
# Apply categorical sorting to Stop Name. 
sort_stops = tu_df_merged[tu_df_merged.trip_id=='14010000592040547']
sort_stops = sort_stops.groupby('stop_id').apply(lambda x: x.loc[x.timestamp.idxmax()])
sort_stops = sort_stops.sort_values('arrival_ts')
print(sort_stops.stop_name.to_list())

# Apply sorting settings to categorical stop id.
tu_df_merged.stop_name = pd.Categorical(tu_df_merged.stop_name, sort_stops.stop_name, ordered=True)
```

```python
sort_stops
```

# Delay

From: https://gtfs.org/realtime/reference/#message-stoptimeupdate

* Positive is a delay
* Negative is ahead of schedule

```python
# EXAMPLE: Delay details of a random trip on a random day at 14:51. 
single = tu_df_merged[tu_df_merged.trip_id == '14010000592040547']
single = single.groupby('stop_id').apply(lambda x: x.loc[x.timestamp.idxmax()])
single = single.reset_index(drop=True)

# Apply sorting system
single.stop_name = pd.Categorical(single.stop_name, sort_stops.stop_name, ordered=True)
single.sort_values('stop_id').head()
```

```python
# Only southbound trips. 
groupall = tu_df_merged[tu_df_merged.direction_id==0.0]

# Group by trip and stop and then sort by timestamp to select latest update
groupall = groupall.groupby(['trip_id', 'stop_id']).apply(
    lambda x: x.loc[x.timestamp.idxmax()] if len(x) > 0 else None)

groupall = groupall.drop(columns=['trip_id', 'stop_id', 'direction_id'])
groupall.tail()
```

```python
delay_stats_0 = groupall.pivot_table(values='departure_delay', columns='trip_id', index='stop_name')
# delay_stats_0 = delay_stats_0.fillna(0)

delay_stats_0 = delay_stats_0.reset_index()
delay_stats_0.stop_name = pd.Categorical(delay_stats_0.stop_name, sort_stops.stop_name, ordered=True)
delay_stats_0 = delay_stats_0.set_index('stop_name').sort_index()

delay_stats_0 = delay_stats_0.drop(columns=delay_stats_0.columns[delay_stats_0.max() > 1000])
delay_stats_0 = delay_stats_0.drop(columns=delay_stats_0.columns[delay_stats_0.notna().sum() < 9])

delay_stats_0.plot(figsize=(20,10), legend=False)

delay_stats_0
```

```python
fig, ax = plt.subplots(figsize=(20,70))
sns.violinplot(data=delay_stats_0.T, ax=ax, orient='h', )
```
