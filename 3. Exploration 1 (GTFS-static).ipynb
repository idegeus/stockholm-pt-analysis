```python
import gtfs_kit as gk
from tqdm.notebook import tqdm
from glob import glob
import pandas as pd

# LINES_NAME = 'camille'
LINES_NAME = 'tvarbanan'
example = './data/0-koda-gtfs-static/2021-06-03.zip'
```

```python
feed = gk.read_feed(example, dist_units='km')
gk.list_feed(example)
```

```python
selected_routes = feed.routes
# selected_routes = feed.routes[feed.routes.route_short_name.isin(['41', '42', '42X'])]
# selected_routes = feed.routes[feed.routes.route_short_name.isin(['2'])]
# selected_routes = selected_routes[selected_routes.route_desc == 'blåbuss']
selected_routes = selected_routes[selected_routes.route_desc == 'Tvärbanan']
selected_routes = selected_routes[selected_routes.route_short_name.isin(['30'])]
selected_routes
```

```python
trips_list = []
for day in tqdm(glob('./data/0-koda-gtfs-static/*.zip')):
    feed = gk.read_feed(day, dist_units='km')
    trips_list.append(feed.trips[feed.trips.route_id.isin(selected_routes.route_id)])
    del feed
```

```python
trips_df = pd.concat(trips_list)
# trips_df = trips_df.drop_duplicates(subset='trip_id')
# trips_df = trips_df.merge(feed.routes, how='left', on='route_id')
# trips_df.to_csv(f"""./data/2-working-data/{LINES_NAME}-trips.csv""")
trips_df
```

```python
trips_df.trip_id.str.contains('14010513453884215').sum()
```

```python
stops = feed.get_stops(route_ids=set(trips_df.route_id))
stops.to_csv(f"""./data/2-working-data/{LINES_NAME}-stops.csv""")
stops
```
